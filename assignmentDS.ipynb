```python
# Check missing values
missing = df.isnull().sum()
print(f"\nNumber of missing values: {missing}")
```

```python
# Get information about the DataFrame after checking for missing values
df.info()
```

```python
# Check the unique values in the 'Effective date' column
df['Effective date'].unique()
```

```python
# Fill missing values in 'Effective date' with the mode (most frequent value)
df['Effective date'] = df['Effective date'].fillna(df['Effective date'].mode()[0])
```

```python
# Get information about the DataFrame after filling missing values in 'Effective date'
df.info()
```

```python
# Check the unique values in the 'Financing Type' column
df['Financing Type'].unique()
```

```python
# Fill missing values in 'Amount Normalized' with the mode (most frequent value)
df['Amount Normalized'] = df['Amount Normalized'].fillna(df['Amount Normalized'].mode()[0])
```

```python
# Get information about the DataFrame after filling missing values in 'Amount Normalized'
df.info()
```

```python
# Display the entire DataFrame after filling missing values
df
```

```python
# Fill missing values in 'Financing Type Normalized' with the mode (most frequent value)
df['Financing Type Normalized'] = df['Financing Type Normalized'].fillna(df['Financing Type Normalized'].mode()[0])
```

```python
# Drop the original 'Financing Type' column
df.drop(columns=['Financing Type'], inplace=True)
```

```python
# Get information about the DataFrame after dropping 'Financing Type'
df.info()
```

```python
# Fill missing values in 'Investors' with the mode (most frequent value)
df['Investors'] = df['Investors'].fillna(df['Investors'].mode()[0])
```

```python
# Fill missing values in 'Investors Count' with the mean
df['Investors Count'] = df['Investors Count'].fillna(df['Investors Count'].mean())
```

```python
# Get information about the DataFrame after filling missing values in 'Investors' and 'Investors Count'
df.info()
```

```python
# Import LabelEncoder for categorical feature encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
# Get information about the DataFrame before encoding
df.info()
```

```python
# Define the list of categorical columns to be encoded
categorical_cols = ['Website Domain', 'Effective date', 'Found At', 'Financing Type Normalized', 'Categories', 'Investors', 'Source Urls']
```

```python
# Apply Label Encoding to the specified categorical columns
for col in categorical_cols:
  df[col] = label_encoder.fit_transform(df[col])
```

```python
# Display the DataFrame after Label Encoding
df
```
